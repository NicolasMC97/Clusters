# Import libraries

```python
import random as rnd
import pandas as pd
```

```python
amount_sections=2
sections=range(amount_sections)

limit_upper_lat=[4.755422,4.678025]
limit_lower_lat=[4.663925,4.553890]
limit_upper_lon=[74.075022,74.135649]
limit_lower_lon=[74.047307,74.087928]
```

```python

lat_list=[]
lon_list=[]
id_data=[]
day=[]

for j in sections:
    for i in range (50):
        lat =rnd.uniform(limit_lower_lat[j],limit_upper_lat[j])
        lon=-rnd.uniform(limit_lower_lon[j],limit_upper_lon[j])
        lat_list.append(lat)
        lon_list.append(lon)
        id_data.append(i)
        day.append(rnd.randint(1,6))

d = {'id':id_data,'lat': lat_list, 'lon': lon_list, 'delivery_day' : day}
df = pd.DataFrame(data=d)
df
```

```python
import pandas as pd
import plotly.express as px

fig = px.scatter_mapbox(df , lat="lat" , lon="lon" , hover_name="id" , hover_data=["id"] ,color= "delivery_day" , zoom = 10 , height = 500,color_continuous_scale=px.colors.cyclical.IceFire)
fig.update_layout(mapbox_style= "carto-positron")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=6, random_state=0, n_init="auto",).fit(df[['lat','lon']])

df['cluster']=kmeans.labels_
```

```python
fig = px.scatter_mapbox(df , lat="lat" , lon="lon" , hover_name="id" , hover_data=["id"] ,color= "cluster" , zoom = 10 , height = 500,color_continuous_scale=px.colors.cyclical.IceFire)
fig.update_layout(mapbox_style="carto-positron")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
lat_list=[]
lon_list=[]
id_data=[]

for j in sections:
    for i in range (1000):
        lat =rnd.uniform(limit_lower_lat[j],limit_upper_lat[j])
        lon=-rnd.uniform(limit_lower_lon[j],limit_upper_lon[j])
        lat_list.append(lat)
        lon_list.append(lon)
        id_data.append(i)
        day.append(rnd.randint(1,6))

d = {'id_data':id_data,'lat': lat_list, 'lon': lon_list}
df_new = pd.DataFrame(data=d)

df_new
```

# Using Kmeans to predict

```python
df_new["new_cluster"]=kmeans.predict(df_new[['lat','lon']])

fig = px.scatter_mapbox(df_new , lat="lat" , lon="lon" , hover_name="id_data" , hover_data=["id_data"] ,color= "new_cluster" , zoom = 10 , height = 500,color_continuous_scale=px.colors.cyclical.IceFire)
fig.update_layout(mapbox_style="carto-positron")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

# Using KNN to predict

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

n_neighbors = 5

X_train, X_test, y_train, y_test = train_test_split(df[['lat','lon']], df["cluster"], random_state=0) 
knn = KNeighborsClassifier(n_neighbors)
knn.fit(X_train, y_train)

print('Accuracy of K-NN classifier on training set: {:.2f}'
     .format(knn.score(X_train, y_train)))
print('Accuracy of K-NN classifier on test set: {:.2f}'
     .format(knn.score(X_test, y_test)))
```

```python
df_new["new_cluster"]=knn.predict(df_new[["lat","lon"]])
fig = px.scatter_mapbox(df_new , lat="lat" , lon="lon" , hover_name="id_data" , hover_data=["id_data"] ,color= "new_cluster" , zoom = 10 , height = 500,color_continuous_scale=px.colors.cyclical.IceFire)
fig.update_layout(mapbox_style="carto-positron")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
