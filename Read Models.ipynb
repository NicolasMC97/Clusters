```python
import random as rnd
import pandas as pd
```

```python
lat_list=[]
lon_list=[]
id_data=[]
dia=[]

for i in range (1000):
    lat =rnd.uniform(4.69,4.61)
    lon=-rnd.uniform(74.09,74.07)
    lat_list.append(lat)
    lon_list.append(lon)
    id_data.append(i)
    dia.append(rnd.randint(1,6))

d = {'id_data':id_data,'lat': lat_list, 'lon': lon_list, 'bloque_dia' : dia}
df = pd.DataFrame(data=d)
df
```

```python
import pandas as pd
import plotly.express as px

fig = px.scatter_mapbox(df , lat="lat" , lon="lon" , hover_name="id_data" , hover_data=["id_data"] ,color= "bloque_dia" , zoom = 11 , height = 500,color_continuous_scale=px.colors.cyclical.IceFire)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=6, random_state=0, n_init="auto",).fit(df[['lat','lon']])

df['cluster']=kmeans.labels_

fig = px.scatter_mapbox(df , lat="lat" , lon="lon" , hover_name="id_data" , hover_data=["id_data"] ,color= "cluster" , zoom = 11 , height = 500,color_continuous_scale=px.colors.cyclical.IceFire)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
lat_list=[]
lon_list=[]
id_data=[]

for i in range (10000):
    lat =rnd.uniform(4.69,4.57)
    lon=-rnd.uniform(74.09,74.14)
    lat_list.append(lat)
    lon_list.append(lon)
    id_data.append(i)

d = {'id_data':id_data,'lat': lat_list, 'lon': lon_list}
df = pd.DataFrame(data=d)

df
```

```python
df["new_cluster"]=kmeans.predict(df[['lat','lon']])

fig = px.scatter_mapbox(df , lat="lat" , lon="lon" , hover_name="id_data" , hover_data=["id_data"] ,color= "new_cluster" , zoom = 11 , height = 500,color_continuous_scale=px.colors.cyclical.IceFire)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
